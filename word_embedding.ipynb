```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = {
    'This is the first new sentence',
    'Here is the second new sentence',
    'A third sentence for the set',
    'Another one to add',
    'The fifth sentence goes here',
    'And the sixth one',
    'Finally, the seventh sentence'
}
```

```python
sent
```

```python
voc_size=1000
```

One Hot Representation


```python
one_hot_rep = [one_hot(words, voc_size) for words in sent]
print(one_hot_rep)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_lenght = 10
embedded_docs=pad_sequences(one_hot_rep, padding='pre', maxlen=sent_lenght)
print(embedded_docs)
```

```python
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, 10, input_length=sent_lenght))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
